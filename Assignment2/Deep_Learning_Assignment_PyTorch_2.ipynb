Q2) a) Write a function to generate a histogram for a grayscale image and plot the original image alongside its histogram. Do it
for all of the 10 chosen images.

```python
def hist_grayscale(img_input):
    # Read the image from the specified path
    img = cv2.imread(img_input)
    
    # Resize the image to a fixed size (500x500)
    img = cv2.resize(img, (500, 500))
    
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Calculate the histogram of the grayscale image
    hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
    
    # Create a figure to display both the grayscale image and its histogram
    plt.figure(figsize=(12, 5))
    
    # Plot the grayscale image on the left
    plt.subplot(1, 2, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')
    
    # Plot the histogram on the right
    plt.subplot(1, 2, 2)
    plt.title("Grayscale Histogram")
    plt.xlabel("Bins")
    plt.ylabel("# of Pixels")
    plt.plot(hist)
    plt.xlim([0, 256])
    
    # Display the image and histogram
    plt.tight_layout()
    plt.show()
    
    # Display the grayscale image using OpenCV
    #cv2.imshow('Grayscale', gray_image)
    
    # Wait for a key press to close the window
   # cv2.waitKey(0)
    
    # Close all windows
    #cv2.destroyAllWindows()
```

b) Write a function to implement histogram equalization in Python. Display both the original and equalized images, and
compare their histograms. Do it for all of the 10 chosen images.

```python
def hist_equalized(img_input):
    # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply histogram equalization
    equ_img = cv2.equalizeHist(gray_img)
    
    # Create a figure to plot both the original and equalized images and their histograms
    plt.figure(figsize=(10, 8))
    
    # Plot original grayscale image and its histogram
    plt.subplot(2, 2, 1)
    plt.imshow(gray_img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    plt.subplot(2, 2, 2)
    hist_orig = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
    plt.plot(hist_orig)
    plt.title('Original Histogram')
    plt.xlim([0, 256])
    
    # Plot equalized image and its histogram
    plt.subplot(2, 2, 3)
    plt.imshow(equ_img, cmap='gray')
    plt.title('Equalized Image')
    plt.axis('off')
    
    plt.subplot(2, 2, 4)
    hist_equ = cv2.calcHist([equ_img], [0], None, [256], [0, 256])
    plt.plot(hist_equ)
    plt.title('Equalized Histogram')
    plt.xlim([0, 256])
    
    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
    
   # cv2.waitKey(0)
    #cv2.destroyAllWindows()
```

Q3) Create a binary image based on a threshold, and display theoriginal image, the thresholded binary image, and the result of
blending the binary map with the original image. Do it for all of the10 chosen images.

```python
def binary_img(img_input):
    # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply binary thresholding
    _, thresholded_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
    
    # Convert the thresholded image to a 3-channel image (so it matches the color image)
    thresholded_img_color = cv2.cvtColor(thresholded_img, cv2.COLOR_GRAY2BGR)
    
    # Plot the original image on the left 
    plt.subplot(1, 3, 1)
    plt.imshow(img)  
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the thresholded binary image in the middle
    plt.subplot(1, 3, 2)
    plt.imshow(thresholded_img, cmap='gray')
    plt.title('Thresholded Binary Image')
    plt.axis('off')
    
    # Blend the original and thresholded images
    blend = cv2.addWeighted(img, 0.5, thresholded_img_color, 0.5, 0.0)
    
    # Plot the blended image on the right
    plt.subplot(1, 3, 3)
    plt.imshow(blend)
    plt.title('Blended Image')
    plt.axis('off')
    
    # Display the plots
    plt.tight_layout()
    plt.show()
```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    binary_img(first_imgs[i])
```

```python
def avg_pooling(img_input):
     # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    # Define a 3x3 averaging kernel
    kernel = np.ones((3, 3), np.float32) / 9
    # Apply convolution using filter2D
    convolved_img = cv2.filter2D(img, -1, kernel)

    # Plot the original image on the left 
    plt.subplot(1, 2, 1)
    plt.imshow(img)  
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the Convolved image on the right
    plt.subplot(1, 2, 2)
    plt.imshow(convolved_img)
    plt.title('Convolved Image')
    plt.axis('off')

    # Display the plots
    plt.tight_layout()
    plt.show()
    
```

```python
def filter_lowpass(img_input):
     # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    # Define kernels separately for Gaussian blur and averaging blur
    # For GaussianBlur, the kernel size is specified (21, 21) with 3.5 as sigma (std deviation)
    gaussian_kernel_size = (21, 21)

    # For Averaging blur, kernel size is defined separately
    average_kernel_size = (21, 21)

    # Apply Gaussian blur with the defined kernel
    gaussian_blur = cv2.GaussianBlur(img, gaussian_kernel_size, 3.5)

    # Apply Averaging blur with the defined kernel
    average_blur = cv2.blur(img, average_kernel_size)

    # Plot the original image on the left 
    plt.subplot(1, 3, 1)
    plt.imshow(img)  
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the Gaussian blurred image in the middle
    plt.subplot(1, 3, 2)
    plt.imshow(gaussian_blur)
    plt.title('With Gaussian filter')
    plt.axis('off')
    
    # Plot the box filtered image on the right
    plt.subplot(1, 3, 3)
    plt.imshow(average_blur)
    plt.title('With Box filter')
    plt.axis('off')
    
    # Display the plots
    plt.tight_layout()
    plt.show()
```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    filter_lowpass(first_imgs[i])
```

```python
def filter_highpass(img_input):
     # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    # Define kernel size for Gaussian blur (low-pass filter)
    gaussian_kernel_size = (3, 3)
    
    # Apply Gaussian blur with the defined kernel
    gaussian_blur = cv2.GaussianBlur(img, gaussian_kernel_size, 1)

    # Subtract the blurred (low-pass) image from the original image to get the high-pass filtered image
    highpass_img = cv2.subtract(img, gaussian_blur)
    
    # Plot the original image on the left 
    plt.subplot(1, 2, 1)
    plt.imshow(img)  
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the highpass filtered image on right
    plt.subplot(1, 2, 2)
    plt.imshow(highpass_img)
    plt.title('With Highpass filter')
    plt.axis('off')
    
    # Display the plots
    plt.tight_layout()
    plt.show()
```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    filter_highpass(first_imgs[i])
```

```python
def sharpening(img_input):
    # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    #defining the kernal for image sharepening
    kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
    ])
    sharpened = cv2.filter2D(img, -1, kernel)
    
     # Plot the original image on the left 
    plt.subplot(1, 2, 1)
    plt.imshow(img)  
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the sharpened image on right
    plt.subplot(1, 2, 2)
    plt.imshow(sharpened)
    plt.title('Sharpened Immage')
    plt.axis('off')
    
    # Display the plots
    plt.tight_layout()
    plt.show()

```

```python
def weiner_filtering(img_input):
    # Read the image
    img = cv2.imread(img_input, cv2.IMREAD_GRAYSCALE)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Create a figure
    plt.figure(figsize=(12, 5))
    
    # Define point spread function for averaging filter
    psf = np.ones((5, 5)) / 25
    
    # Add Gaussian noise to the image
    noisy_img = img + 0.1 * img.std() * np.random.standard_normal(img.shape)

    # Wiener filtered image
    wiener_img = restoration.wiener(noisy_img, psf, balance=0.1)
    
    # Plot the original image on the left 
    plt.subplot(1, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    # Plot the noisy image in the middle
    plt.subplot(1, 3, 2)
    plt.imshow(noisy_img, cmap='gray')
    plt.title('Noisy Image')
    plt.axis('off')
    
    # Plot the Wiener filtered image on the right
    plt.subplot(1, 3, 3)
    plt.imshow(wiener_img, cmap='gray')
    plt.title('Wiener Filtered Image')
    plt.axis('off')
    
    # Display the plots
    plt.tight_layout()
    plt.show()


```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    weiner_filtering(first_imgs[i])
```

```python
#pass the imagethrough a lowpass filter followed by a high pass filter and observethe difference between the original and the
#obtained image
def mix_high_low(img_input):
    # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image
    img = cv2.resize(img, (500, 500))
    
    # Convert from BGR to RGB format for display in matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Apply Gaussian blur with a large kernel for the initial low-pass filter
    gaussian_kernel_size_large = (21, 21)
    img2 = cv2.GaussianBlur(img, gaussian_kernel_size_large, 3.5)
    
    # Apply Gaussian blur with a smaller kernel (for high-pass filter)
    gaussian_kernel_size_small = (3, 3)
    gaussian_blur = cv2.GaussianBlur(img2, gaussian_kernel_size_small, 1)
    
    # Subtract the blurred image to get the high-pass filtered image
    highpass_img = cv2.subtract(img2, gaussian_blur)
    
    # Convert the highpass image to RGB for displaying in matplotlib
    highpass_img_rgb = cv2.cvtColor(highpass_img, cv2.COLOR_BGR2RGB)
    
    # Create a figure and display both original and processed images
    plt.figure(figsize=(12, 5))

    # Display the original image
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')
    
    # Display the high-pass filtered image
    plt.subplot(1, 2, 2)
    plt.imshow(highpass_img_rgb)
    plt.title('Low-pass followed by High-pass Image')
    plt.axis('off')
    
    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    mix_high_low(first_imgs[i])
```

Q5) Calculate the 2D Discrete Fourier Transform(DFT) for each of the 10 images. Now on these images, use the inverse DFT and 
observe how the reconstructed images look. Also calculate the error between the reconstructed and the original images.

```python
def Inverse_and_dft2D(img_input):
    # Read the image
    img = cv2.imread(img_input)
    
    # Resize the image for consistency
    img = cv2.resize(img, (500, 500))

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Perform DFT and shift zero-frequency component to the center
    dft = cv2.dft(np.float32(gray), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    
    # Compute the magnitude spectrum of the Fourier Transform
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

    # Normalize the magnitude for display
    magnitude_spectrum = cv2.normalize(magnitude_spectrum, None, 0, 255, cv2.NORM_MINMAX)

    # Create a figure to display original image and DFT
    plt.figure(figsize=(12, 6))

    # Plot the original resized image
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')

    # Plot the magnitude spectrum (DFT result)
    plt.subplot(1, 3, 2)
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title('Magnitude Spectrum (DFT)')
    plt.axis('off')

    # Perform inverse shift to bring the zero-frequency component back
    dft_ishift = np.fft.ifftshift(dft_shift)

    # Compute the inverse DFT
    img_back = cv2.idft(dft_ishift)
    img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])

    # Normalize the reconstructed image for display
    img_back = cv2.normalize(img_back, None, 0, 255, cv2.NORM_MINMAX)

    # Plot the reconstructed image from IDFT
    plt.subplot(1, 3, 3)
    plt.imshow(img_back, cmap='gray')
    plt.title(f'Reconstructed Image(IDFT), mse={np.mean((gray - img_back) ** 2)}')
    plt.axis('off')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
#This will display all 10 images, output by above function
for i in range(0,10):
    Inverse_and_dft2D(first_imgs[i])
```
